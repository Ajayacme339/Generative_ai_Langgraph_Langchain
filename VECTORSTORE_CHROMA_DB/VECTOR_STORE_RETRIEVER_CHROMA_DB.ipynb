#### Chroma
Chroma is a AI-native open-source vector database focused on developer productivity and happiness. Chroma is licensed under Apache 2.0.

https://python.langchain.com/v0.2/docs/integrations/vectorstores/

```python
## build a simple vector database with Chroma
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OpenAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python



loader = TextLoader("speech.txt")
data = loader.load()
data
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
splits = text_splitter.split_documents(data)
splits
```

```python
embeddings = OpenAIEmbeddings()
embeddings
```

```python
vector_db = Chroma.from_documents(splits, embedding=embeddings)
vector_db
```

```python
query = "What does the speaker believe is the main reason the United States should enter the war?"
print(query)
```

```python
query_result = vector_db.similarity_search(query)
query_result[0].page_content
```

```python
### Saving to disk
vector_db = Chroma.from_documents(splits, embedding=embeddings,persist_directory="./chroma_db")
```

```python
#load from disk 



db2 = Chroma(persist_directory="./chroma_db",embedding_function=embeddings)
docs=db2.similarity_search(query)
print(docs[0].page_content)   
```

```python
#### Similarirty with a Search Score 
docs = vector_db.similarity_search_with_score(query)
docs
```

```python
##### Retrieval 



retrievar = vector_db.as_retriever()
retrievar.invoke(query)[0].page_content
```
